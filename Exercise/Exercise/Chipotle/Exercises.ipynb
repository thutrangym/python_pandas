### Step 1. Import the necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter

# set this so the graphs open internally
%matplotlib inline
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/chipotle.tsv).

```python
file_path = r"D:\Download\Exercise\Exercise\07_Visualization\Chipotle\chipotle.tsv"
data = pd.read_csv(file_path, sep="\t")
```

### Step 3. Assign it to a variable called chipo.

```python
chipo = pd.DataFrame(data)
```

### Step 4. See the first 10 entries

```python
print(chipo.head(10))
```

### Step 5. Create a histogram of the top 5 items bought

```python
top_items = chipo['item_name'].value_counts().head(5)
print(top_items)

plt.plot(top_items.index, top_items.values, color='blue')
plt.title('Top 5 Items Ordered')
plt.xlabel('Items') 
plt.ylabel('Number of Orders')
plt.show()
```

### Step 6. Create a scatterplot with the number of items orderered per order price
#### Hint: Price should be in the X-axis and Items ordered in the Y-axis

```python
plt.figure(figsize=(10, 6))
plt.scatter(chipo['item_price'], chipo['quantity'], alpha=0.5, color='blue')
plt.title('Item Price vs Quantity Ordered') 
plt.xlabel('Item Price')    
plt.ylabel('Quantity Ordered')
```

### Step 7. BONUS: Create a question and a graph to answer your own question.

```python
# Plotting the distribution of order quantities
plt.figure(figsize=(10, 6))
chipo['quantity'].value_counts().sort_index().plot(kind='bar', color='orange')
plt.title('Distribution of Order Quantities')
plt.xlabel('Quantity Ordered')
plt.ylabel('Number of Orders')
plt.xticks(rotation=0)
plt.show()
```
